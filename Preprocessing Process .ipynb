```python
from SummarizationOptimizerEnglish import *
from IPython import display

import copy
```

```python
dataframe = pd.read_csv("english_news.csv") #Data Training/Ringkasan Pakar
dataframe = dataframe.dropna()
topik = dataframe["Topik"].unique()
```

```python
def copy_class(ft1,setType):
  ft2 = copy.copy(ft1)
  ft2.tfidf_compress = setType
  ft2.df_fiks = ft2.Fitur8(ft2.df_clean)
  return ft2
```

```python
import os
import pickle

treshhold = True
tfidf     = True
for i,d in enumerate(dataframe["Topik"].unique()):
  # if i > 98:
    data = dataframe[dataframe["Topik"]==d].copy()
    print("Jalan Dokumen",i,len(data))
    df_clean = Preparation_Data(data).df_clean
    extrackFitur = {}
    extrackFitur["standard"] = FeatureExtraction(df = df_clean,tfidf=True,tfidf_compress="",cleaning=False)
    extrackFitur["sum"]      = copy_class(extrackFitur["standard"],"sum")
    extrackFitur["mean"]     = copy_class(extrackFitur["standard"],"mean")
    for sin_conv in [True,False]:
      for markov in [True,False]:
        for tfidf_compress in ["sum","mean",""]:

          ext = tfidf_compress
          if tfidf_compress == "":
            ext = "standard"

          print("Jalan Proses",ext,f"\n{i}_sin_conv_{sin_conv}_markov_{markov}_tfidf_compress_{tfidf_compress}")
          if f"{i}_sin_conv_{sin_conv}_markov_{markov}_tfidf_compress_{tfidf_compress}.pickle" not in os.listdir("PreprocessingEnglish"):
            preprocess = PreprocessTuna(df=data, sin_conv=sin_conv,treshhold = treshhold,markov = markov,tfidf = tfidf,tfidf_compress = tfidf_compress, extrackFitur = extrackFitur[ext])
            with open(f"PreprocessingEnglish/{i}_sin_conv_{sin_conv}_markov_{markov}_tfidf_compress_{tfidf_compress}.pickle", 'wb') as outfile:
              pickle.dump(preprocess, outfile)
    display.clear_output()
```
